```python
from bs4 import BeautifulSoup
import requests
```

```python
URL = "https://books.toscrape.com/"
html = requests.get(URL)
```

```python
URL_source_code=html.text
URL_source_code           
```

```python
soup=BeautifulSoup(URL_source_code,"html.parser")
soup
```

```python
soup.find("div", class_="product_price")
```

```python
soup.find("div", class_="product_price").find("p", class_="price_color").text
```

```python
while True:
    for i in soup.find_all(class_="product_price"): #pay attention to the find_all
        print(i.find("p", class_="price_color").text)
    
    nextlink = soup.find(class_="next")
    
    if nextlink:
        nextlink = URL + nextlink.a['href']
        subhtml = requests.get(nextlink)
        soup = BeautifulSoup(subhtml.text, "html.parser")
    else:
        break
```

```python
price_list=[]
for number in range(3):
    #get URL
    URL="http://books.toscrape.com/catalogue/page-"+str(number+1)+str(".html") #why did we do number +1?
    html=requests.get(URL)
    
    #get URL source code
    URL_source_code=html.text
    
    #parse the source code
    soup=BeautifulSoup(URL_source_code,"html.parser")

    for i in soup.find_all("div", class_="product_price"): 
    
        price_list.append(i.find("p", class_="price_color").text)
    
        
print(price_list)
```

```python
_re_digits = re.compile(r"(-?(?:(?:\d+(?:\.\d*)?)|(?:\.\d+)))")
digits = []
for element in price_list:
    digits += [ float(n) for n in _re_digits.findall(element)]
    
digits
```
